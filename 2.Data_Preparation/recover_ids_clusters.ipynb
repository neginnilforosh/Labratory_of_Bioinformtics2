```python
import pandas as pd

# read only the first column to get the ids
col1_neg = pd.read_csv("neg_cluster/neg_cluster.tsv", sep="\t", header=None, usecols=[0], dtype=str, names=["id"])
col1_neg["id"] = col1_neg["id"].str.strip()  # remove spaces and newline characters
col1_pos = pd.read_csv("pos_cluster/pos_cluster.tsv", sep="\t", header=None, usecols=[0], dtype=str, names=["id"])
col1_pos["id"] = col1_pos["id"].str.strip()  # remove spaces and newline characters
```

```python
#checking the negative ids
col1_neg.head(10)
```

```python
#checking the positive ids
col1_pos.head(10)
```

```python
# unique values of the first column of negatives (to take the representative ids only one time)
unique_ids_neg = col1_neg["id"].drop_duplicates()
unique_ids_neg.head(10)
unique_ids_list_neg=unique_ids_neg.values.tolist()

# unique values of the first column of positives (to take the representative ids only one time)
unique_ids_pos = col1_pos["id"].drop_duplicates()
unique_ids_pos.head(10)
unique_ids_list_pos=unique_ids_pos.values.tolist()
```

```python
#recovering of the entire negative dataset
col2_neg = pd.read_csv("../1.Data_Collection/negative_dataset.tsv", sep="\t")
col2_neg.head(10)
```

```python
#recovering of the entire positive dataset
col2_pos = pd.read_csv("../1.Data_Collection/positive_dataset.tsv", sep="\t")
col2_pos.head(10)
```

```python
#Extraction negative representative proteins data and creating of the tsv file
intestation=("EntryID","OrganismName","Kingdom","SequenceLength","HelixDomain")
output_neg_file_name='negative_representative_data.tsv'
with open(output_neg_file_name, "w") as ofs:
        print(*intestation, sep="\t", file=ofs)
        for index,row in col2_neg.iterrows():
          if row['EntryID'] in unique_ids_list_neg:
            print(*row,sep="\t",file=ofs)
        ofs.close

```

```python
#Extraction positive representative proteins data and creating of the tsv file
intestation=("EntryID","OrganismName","Kingdom","SequenceLength","SPStart","SPEnd")
output_pos_file_name='positive_representative_data.tsv'
with open(output_pos_file_name, "w") as ofs:
        print(*intestation, sep="\t", file=ofs)
        for index,row in col2_pos.iterrows():
          if row['EntryID'] in unique_ids_list_pos:
            print(*row,sep="\t",file=ofs)
        ofs.close
```
